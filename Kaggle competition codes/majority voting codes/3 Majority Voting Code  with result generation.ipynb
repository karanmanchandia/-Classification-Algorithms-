```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
col_names = [i for i in range(101)]
dataset = pd.read_csv("train_features.csv",names = col_names)

pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', 1000)
```

```python
# Removing first col
X_train = dataset.iloc[:, 1: ].values
```

```python
col_names1 = [i for i in range(2)]
labels = pd.read_csv("train_label.csv",names = col_names1)
```

```python
y_train = labels.iloc[1:, -1].values
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
temp1 = [i for i in range(101)]
test_data = pd.read_csv("test_features.csv",names = temp1)
X_test = test_data.iloc[:, 1: ].values
```

```python
X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='poly', gamma=10, degree=2)
from sklearn.ensemble import BaggingClassifier
bag_model1=BaggingClassifier(base_estimator=svclassifier,n_estimators=100,bootstrap=True)
bag_model1=bag_model1.fit(X_train,y_train)
y_pred_bagsvm=bag_model1.predict(X_test)
y_pred_bagsvm = y_pred_bagsvm.astype(int)
y_pred_bagsvm = y_pred_bagsvm.tolist()
print(y_pred_bagsvm)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9, weights='distance')
from sklearn.ensemble import BaggingClassifier
bag_model2=BaggingClassifier(base_estimator=knn,n_estimators=100,bootstrap=True)
bag_model2=bag_model2.fit(X_train,y_train)
y_pred_bagknn=bag_model2.predict(X_test)
y_pred_bagknn = y_pred_bagknn.astype(int)
y_pred_bagknn = y_pred_bagknn.tolist()
print(y_pred_bagknn)
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=10, multi_class='ovr', C=0.1, penalty='l1')
# print(LR)
LR.fit(X_train,y_train)
y_pred_lr = LR.predict(X_test)
y_pred_lr = y_pred_lr.astype(int)
print(y_pred_lr)
y_pred_lr = y_pred_lr.tolist()
print(y_pred_lr)
print(type(y_pred_lr))
print(len(y_pred_lr))
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=20, random_state=8, max_features=10)
regressor.fit(X_train, y_train)
y_pred1 = regressor.predict(X_test)
print(y_pred1)
print(type(y_pred1))

y_pred_rf = []
for item in y_pred1:
    temp = round(item)
    y_pred_rf.append(temp)

y_pred_rf = np.asarray(y_pred_rf)
y_pred_rf = y_pred_rf.astype(int)
print(y_pred_rf)

y_pred_rf = y_pred_rf.tolist()
print(y_pred_rf)
print(type(y_pred_rf))
```

```python
from sklearn.ensemble import AdaBoostClassifier
abc =AdaBoostClassifier(n_estimators=9, learning_rate=0.5)
from sklearn.ensemble import BaggingClassifier
bag_model3=BaggingClassifier(base_estimator=abc,n_estimators=100,bootstrap=True)
bag_model3=bag_model3.fit(X_train,y_train)
y_pred_bagada=bag_model3.predict(X_test)
y_pred_bagada = y_pred_bagada.astype(int)
y_pred_bagada = y_pred_bagada.tolist()
print(y_pred_bagada)
```

```python
df = pd.DataFrame(list(zip(y_pred_bagsvm, y_pred_lr,y_pred_bagknn, y_pred_rf, y_pred_bagada)), index =[i for i in range(418,796)])
```
